```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import silhouette_score as sil

nc=[2,3,4,5,6,7,8,9]

for i in nc:
    for linkages in ['ward','single','average','complete']:
        hc = AgglomerativeClustering(n_clusters=i,affinity = 'euclidean', linkage = linkages)
        y = hc.fit_predict(df_crime)
        sc=sil(df_crime,y)
        print('Clusters = {}, linkage = {}, Sc score = {}'.format(i,linkages,sc))
    print()
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_samples,silhouette_score
```

```python
crimedataa=pd.read_csv('crime_data.csv')
crimedataa
```

```python
crimedataa.rename({"Unnamed: 0":"states"},axis=1)
```

```python
crimedata01=crimedataa.iloc[:,1:]
crimedata01
```

```python
#scaler=StandardScaler()
#crime_scaled=scaler.fit_transform(crime.iloc[:,1:5])
#x=crime_scaled
#normailising the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
x=norm_func(crimedata01.iloc[:,1:])
```

```python
nc=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=10)
    kmeans.fit(x)
    nc.append(kmeans.inertia_)
    
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.plot(range(1,11),nc)
plt.scatter(range(1,11),nc,color='red')
plt.xticks(range(1,11))
plt.title('Elbow graph')
```

```python
clustered=KMeans(n_clusters=4,init='k-means++',random_state=10)
clustered.fit(x)
```

```python
clustered.labels_
```

```python
crimedataa['k-means_cluster']=clustered.labels_
```

```python
crimedataa
```

```python
for i in np.array([0,1,2,3]):
    print('States that belong to cluster index',i)
    print(crime[crime['k-means_cluster']==i])
    print('\n')
```

```python
crime003=crime.sort_values('k-means_cluster')
crime003.iloc[:,[-1]]
```

```python
color=[]
for i in crimedataa['k-means_cluster']:
        if i==0:
            color.append("red")
        elif i==1:
            color.append("black")
        elif i==2:
            color.append("blue")
        elif i==3:
            color.append("orange")   
plt.scatter(crime.iloc[:,2],crime.iloc[:,0],c=color)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crimedata004=pd.read_csv('crime_data.csv')
crimedata004
```

```python
crime005=crimedata004.rename({'Unnamed: 0':"state"},axis=1)
crime005.head()
```

```python
crimedata006=crime005.drop(['state'],axis=1)
crimedata006.head()
```

```python
norm_crime=StandardScaler().fit(crimedata006)
X=norm_crime.transform(crimedata006)
```

```python
X
```

```python
dbscan=DBSCAN(eps=1,min_samples=3)
dbscan.fit(X)
```

```python
#noisy samples are given the lable as -1
dbscan.labels_
```

```python
crime005['dbscan.labels_']=dbscan.labels_
```

```python
crime005
```

```python
crime005.mean()
```

```python
crimedf=crime005.sort_values('dbscan.labels_')
crimedf
```

```python
#Total Four clusters obtained.
```

```python
color=[]
for i in crimedf['dbscan.labels_']:
        if i==-1:
            color.append("pink")
        if i==0:
            color.append("red")
        elif i==1:
            color.append("black")
        elif i==2:
            color.append("blue")
        elif i==3:
            color.append("orange")   
plt.scatter(crimedf.iloc[:,0],crimedf.iloc[:,2],c=color)
```
